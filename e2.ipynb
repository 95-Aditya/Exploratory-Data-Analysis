```python
plt.bar(e2["Country"].loc[:20],e2["Local Purchasing Power Index"].loc[:20],color=(e2["Local Purchasing Power Index"].loc[:20]>100.0).map({True:"red",False:"green"}),width=0.8)
plt.gcf().set_size_inches(14,7)
plt.xlabel("Countries")
plt.ylabel("Local Purchasing Power Index")
plt.xticks(rotation=90)
plt.title("Top 20 Countries")
plt.show()
```

Local Purchasing Power: This index indicates the relative purchasing power in a given city based on the average net salary. A domestic purchasing power of 40 means that residents with an average salary can afford, on average, 60% less goods and services compared to residents of New York City with an average salary.

By above graph(for top 20 countries), Luxembory is having highest Local Purchasing Power Index.This means the residents of luxemborg with an average salary need to afford 82.5% more on goods and services when compared to New York city residents.

Bahamas,Barbodas are having lowest Local Purchasing Power Index in top 20 countries.


```python
plt.bar(e2["Country"].loc[101:120],e2["Local Purchasing Power Index"].loc[101:120])
plt.title("Bottom 20 Countries")
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(e2["Country"].loc[:20],e2["Cost of Living Index"].loc[:20],color='#E4080A',label="Cost of Living Index")
plt.bar(e2["Country"].loc[:20],e2["Cost of Living Plus Rent Index"].loc[:20],color='#DFC57B',label="Cost of Living Plus Rent Index")
plt.xticks(rotation=90)
plt.gcf().set_size_inches(14,7)
plt.legend()
plt.show()
```

Cost of Living Index (Excl. Rent): This index indicates the relative prices of consumer goods like groceries, restaurants, transportation, and utilities. It excludes accommodation expenses such as rent or mortgage. For instance, a city with a Cost of Living Index of 120 is estimated to be 20% more expensive than New York City (excluding rent).
Cost of Living Index is less in most of the countries except for switzerland which is having more than 1.1% more compared to New york city.

Cost of Living Plus Rent Index: This index estimates consumer goods prices, including rent, in comparison to New York City.
Cost of Living Plus Rent Index in the Top 20 countries are low when compared to new york city.
