 I have taken above 300 km range ev's as per my considerations.

```python
e3_r=e3[e3["Range_Km"] >= 300]
e3_r
```

```python
plt.bar(e3_r["Model"],e3_r["Range_Km"],width=0.5,color='#AD1295')
plt.xlabel("Model")
plt.ylabel("Range_Km")
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(e3_r["Model"],e3_r["TopSpeed_KmH"],width=0.5,color='#ADC31A')
plt.xlabel("Model")
plt.ylabel("TopSpeed_KmH")
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(e3_r["Model"],e3_r["AccelSec"],width=0.5,color=(e3_r["AccelSec"] < 4.0).map({True:"#7DDA58",False:"#DEBC5B"}))
plt.xlabel("Model")
plt.ylabel("AccelSec")
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation=90)
plt.show()
```

As per above bar graph acceleration of ev's (0-100)km/hr in less than 4 seconds are Lucid Air, Taycan Turbo S, e-tron GT, Model-3 Long Range Performance,
 CyberTruck Tri Motor, Model S Long Range, Roadster, Model Y Long Range Performance, Taycan Cross Turismo, Taycon Turbo, Model X Performance.

```python
plt.bar(e3_r["Model"],e3_r["PriceEuro"],width=0.5,color="#D95A5C")
plt.xlabel("Model")
plt.ylabel("PriceEuro")
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation=90)
plt.show()
```

As per data based on 3 parmeters which i have considered(top speed,range(above 300 km) and acceleration in sec(0-100km))

Tesla roadster wins big when compared to others but this model of tesla is also most expensive car.
